```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle

%matplotlib inline
```

```python
with open('wasabi_cj_stats.pkl', 'rb') as f:
    stats = pickle.load(f)
```

```python
df = pd.DataFrame(stats) / 10**8
df.head()
```

```python
df.describe()
```

```python
fig,ax = plt.subplots(len(df.columns), figsize=(20, 80))

for index, attr in enumerate(df.columns):
    ax[index].plot(df[attr], label=attr)
    ax[index].legend(loc='best')
#plt.plot(df['min'])
#plt.plot(df['mean'])
```
